### Importing Libraries

```python
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

### Data Preparation

```python
df = pd.read_csv("Dataset/xray_knees.csv")
df
```

```python
df.drop("dcm", axis=1, inplace=True)
```

```python
df['type'].replace(["1-st stage", "2-nd stage", "3-rd stage", "4-th stage"], ["1st stage", "2nd stage", "3rd stage", "4th stage"], inplace=True)
df
```

```python
data_gen = ImageDataGenerator(
    rescale=1./255, rotation_range=20, 
    width_shift_range=0.2, height_shift_range=0.2, shear_range=0.2,
    zoom_range=0.2, horizontal_flip=True, fill_mode='nearest'
)
```

```python
# Generate data batches
batch_size = 32

train_generator = data_gen.flow_from_directory('/kaggle/input/knee-x-rays-4-stages-of-osteoarthritis/files',
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)
```
